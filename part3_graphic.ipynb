```python
#문제1)  평균 구하기

#다음과 같은 학생들의 수학 점수가 있다. 평균 점수를 구하시오.

import pandas as pd

data = {'이름': ['철수', '영희', '민수', '지영', '수지'],
        '수학': [88, 92, 75, 95, 80]}
df = pd.DataFrame(data)
df['수학'].mean()
#df.corr(numeric_only=True)
```

```python
#문제2) 중앙값 구하기
#다음 데이터에서 키의 중앙값을 구하시오.

data = {'이름': ['A', 'B', 'C', 'D', 'E', 'F'],
        '키': [170, 165, 180, 175, 160, 185]}
df = pd.DataFrame(data)
df['키'].mean()
```

```python
#문제3) 최빈값 구하기
#다음 고객 데이터에서 구매한 상품 수의 최빈값을 구하시오.

data = {'고객ID': [101, 102, 103, 104, 105, 106, 107],
        '구매_상품수': [2, 3, 3, 2, 4, 2, 3]}
df = pd.DataFrame(data)
#최빈값 == 원소 빈도가 많은 것 구하기 2-3개, 3-4개, 4-1개
df['구매_상품수'].mode()
```

```python
#문제4) 분산과 표준편차
#다음 데이터를 사용해서 몸무게의 분산과 표준편차를 각각 구하시오.

data = {'이름': ['홍길동', '임꺽정', '장길산', '이순신'],
        '몸무게': [65, 70, 75, 80]}
df = pd.DataFrame(data)
df['몸무게'].mode()
df['몸무게'].var()
df['몸무게'].std()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Jupyter Notebook이라면 그래프 바로 보이게:
# %matplotlib inline

# 예시 데이터
np.random.seed(0)
df = pd.DataFrame({
    '월': ['1월', '2월', '3월', '4월'],
    '매출': [250, 270, 300, 280],
    '지출': [200, 230, 250, 240]
})  # 코드 인위적으로 바꾸지말고 수정만해줘

# 데이터프레임 (size 모두 100으로 통일)
df_random = pd.DataFrame({
    '점수': np.random.normal(loc=70, scale=10, size=100),
    '키': np.random.normal(170, 10, size=100),
    '몸무게': np.random.normal(65, 15, size=100)
})

df_random
```

```python
# 선 그래프 그리기
# 한글 폰트 오류해결 - 파일다운 malgun.ttf 올린다.
from matplotlib import font_manager, rc
font_path = './data/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

df.plot(x='월', y=['매출', '지출'], kind='line', marker='o')
plt.title('월별 매출 및 지출 추이')
plt.ylabel('금액')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Jupyter Notebook이라면 그래프 바로 보이게:
%matplotlib inline

# 예시 데이터
np.random.seed(0)
df = pd.DataFrame({
    '월': ['1월', '2월', '3월', '4월'],
    '매출': [250, 270, 300, 280],
    '지출': [200, 230, 250, 240]
})

df_random = pd.DataFrame({
    '점수': np.random.normal(loc=70, scale=10, size=100),
    '키': np.random.normal(170, 10, 100),
    '몸무게': np.random.normal(65, 15, 100)
})

df_random.plot(kind='hist', alpha=0.7, figsize=(8, 4), grid=True, edgecolor='black')
plt.title('랜덤 데이터 분포 (점수, 키, 몸무게)')
plt.xlabel('값')
plt.ylabel('빈도수')
plt.show()
```

```python
df_sum = df[['매출', '지출']].sum()
df_sum.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('전체 매출/지출 비율')
plt.ylabel('')
plt.show()
```

```python
df_random.plot(kind='scatter', x='키', y='몸무게', alpha=0.5)
plt.title('키 vs 몸무게 산점도')
plt.show()
```

```python
df_random.plot(kind='hexbin', x='키', y='몸무게', gridsize=25, cmap='viridis')
plt.title('키 vs 몸무게 고밀도 산점도')
plt.show()
```
