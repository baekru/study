```python
import pandas as pd

# auto-mpg.csv 파일 불러오기 (이 파일에 'weight' 컬럼이 있음)
df = pd.read_csv('./data/auto-mpg.csv', header=None)

# 컬럼명 지정 (weight 포함!)
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
              'acceleration', 'model year', 'origin', 'name']

# 선생님 코드 그대로 실행
print(df.mean(numeric_only=True))
print(df['weight'].mean())  # ✅ 산술평균 (정상작동)
print(df.weight.mean())     # ✅ 점 접근도 가능

```

```python
# 평균
df.mean(numeric_only=True)
df.mean(numeric_only=True)
df['weight'].mean()     # 산술평균
df.weight.mean()

# 중앙값
df.median(numeric_only=True)
df['origin'].median()

# 표준편차
df.std(numeric_only=True)  # 오타: numric_only → numeric_only
df[['displacement', 'weight', 'acceleration']].std()  # 괄호 닫힘 오류 수정

df.corr(numeric_only=True)
df[['displacement','cylinders']].corr()
```

```python
# ict_수정.xls 파일을 읽어오기
# 데이터프레임을 전치하기( 행열 위치 바꾸기)
# 데이터 정보
# 데이터 형태 보기
# 데이터 요약 보기
# 각 열의 데이터 갯수 보기
# '전문대계'열의 고유값 개수 구하기
# 전체 산술평균 구하기
# '대학원ICT학과' '대학원ICT관련학과'의 산술 평균 구하기
# 전체 중앙값 구하기
# 최대값 구하기
# 최소값 구하기
# 표준편차 구하기
# '대학원ICT학과' '대학원ICT관련학과' 표준편차 구하기
# 전체 상관 계수 구하기
# 상관계수 절대값이 0.5이상인 것들의 열 이름을 리스트로 저장하기
```

```python
# ict_수정.xlsx 파일을 읽어오기
df2 = pd.read_excel('./data/ict_수정.xlsx')
#컬럼명 바꾸기
df2=df2.rename(columns={'학과별(1)':'학과별', '응답자수 (명)':'응답자수' } )
df2
df2 = df2.set_index('학과별')
df2
# 데이터프레임을 전치하기( 행열 위치 바꾸기)
df2 = df2.T
df2
# 데이터 정보
df2.info()
# 데이터 형태 보기
df2.shape
# 데이터 요약 보기
df2.describe()
# 각 열의 데이터 갯수 보기
df2.count()
# '전문대계'열의 고유값 개수 구하기
df2['전문대계'].value_counts()
# 전체 산술평균 구하기
df2.mean()
# '대학원ICT학과' '대학원ICT관련학과'의 산술 평균 구하기
df2[['대학원ICT학과', '대학원ICT관련학과']].mean()
# 전체 중앙값 구하기
df2.median()
# 최대값 구하기
df2.max()
# 최소값 구하기
df2.min()
# 표준편차 구하기
df2.std()
# '대학원ICT학과' '대학원ICT관련학과' 표준편차 구하기
df2[['대학원ICT학과', '대학원ICT관련학과']].std()
df2.corr()
```

```python
# 전체 상관 계수 구하기
import numpy as np
df2_corr = df2.corr()
index_size = len(df2_corr.index) #행의 갯수 16개
col_size = len(df2_corr.columns) # 열의 개수 16개

for i in range(0, index_size): 
    for j in range( 0, col_size):
        if np.abs( df2_corr.iloc[i,j]  ) >= 0.5 :
            print( df2_corr.iloc[i,j] )
```

```python
# 문제 상관계수 값이 절대값 0.5 이상인 것을 열들을 리스트 저장하기
import numpy as np 
df2_corr = df2.corr()
# 상관계수값을 가져와서 0.5이상인지 판단 
# 행수
df2_index_len = len(df2_corr.index) #행의 갯수
# 열수 
df2_col_len = len(df2_corr.columns) 
# 열명들을 리스트에 담았다
df2_col_name  = df2_corr.columns
result_list = []
for i in range(df2_index_len):  # i = 0, 1,2,..15
    for j in range(df2_col_len): # j = 0,1,2,.. 15
       if( (np.abs( df2_corr.iloc[i][j] ) >= 0.5) and (np.abs( df2_corr.iloc[i][j] ) != 1)  ):
           #print( df2_corr.iloc[i][j] )
           result_list.append( df2_col_name[j] )
    print('\n')
# 중복제거하기 위해서 set() 넣기
result_list = set(result_list)
# set자료를 list로 만들기
result_list = list(result_list)
result_list
```
