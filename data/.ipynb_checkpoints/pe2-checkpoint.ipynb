```python
import os
import shutil

files_to_move = [
    "df_sample.csv",
    "df_sample_no_index.csv",
    "df_sample.xlsx",
    "df_sample.json"
]

target_folder = "data"

# í´ë”ê°€ ì—†ë‹¤ë©´ ìƒì„±
os.makedirs(target_folder, exist_ok=True)

# í˜„ì¬ í´ë” ì •ë³´ ì¶œë ¥
print(f"í˜„ì¬ ì‘ì—… í´ë”: {os.getcwd()}")
print(f"ëŒ€ìƒ í´ë” ê²½ë¡œ: {os.path.abspath(target_folder)}\n")

# ì´ë™ ì‹¤í–‰
for file in files_to_move:
    src = file.strip()
    dst = os.path.join(target_folder, src)

    if os.path.exists(src):
        try:
            shutil.move(src, dst)
            print(f"âœ… {src} â†’ {dst}")
        except Exception as e:
            print(f"âš ï¸ {src} ì´ë™ ì¤‘ ì˜¤ë¥˜: {e}")
    else:
        print(f"âŒ {src} ì—†ìŒ â€” ê±´ë„ˆëœ€")
```
